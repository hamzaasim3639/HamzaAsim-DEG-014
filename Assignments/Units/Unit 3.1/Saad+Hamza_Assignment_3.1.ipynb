```python
def label_encoder_v5(path):
    df = pd.read_csv(path)
    df = df.dropna()
    col_names = []
    
    # Parses for categorical columns
    for col in df.columns:
        if df[col].dtypes == 'object':
            col_names.append(col)
            
    print("These are the columns with categorical data: ", col_names)

    while True:
        # Asks user for desired column name to encode
        user_input = input("\nSelect the column(s) you want to encode (press q when done): ")
        if user_input == 'q':
            break

        # Check if user input is a valid column name
        assert user_input in col_names, f"{user_input} is not a categorical column or is an invalid column."

        # Get unique values 
        unique_values = df[user_input].unique()
        
        print(f"\nThese are the variables in {user_input} column: {', '.join(unique_values)}")
        col_map = {}
        
        # Prompt user for order
        for val in unique_values:
            order = int(input(f"Assign the order to '{val}': "))
            col_map[val] = order

        # Create new column with label encoded values
        new_col_name = f"new_{user_input}_label"
        df[new_col_name] = df[user_input].map(col_map)

        # Drop original column
        df = df.drop(user_input, axis=1)

    return df
```

```python
label_encoder_v5('Athlete_dataset/athlete_events.csv')
```
