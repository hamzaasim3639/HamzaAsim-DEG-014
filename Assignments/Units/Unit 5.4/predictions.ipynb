```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet('./output_data/employee_earnings/earnings_date=2022-02-14/employee_earnings.parquet')
df
```

```python
df["earnings"].mean()
```

```python
lower_bound = df["earnings"].min()
lower_bound
```

```python
upper_bound = df["earnings"].max()
upper_bound
```

```python
np.random.seed(14)
day_6_earnings = np.random.randint(lower_bound, upper_bound, 100)
day_6_earnings
```

```python
np.random.seed(11)
day_7_earnings = np.random.randint(lower_bound, upper_bound, 100)
day_7_earnings
```

```python
day_6_employee_earnings = df.copy()
day_6_employee_earnings["earnings"] = day_6_earnings
day_6_employee_earnings
```

```python
day_7_employee_earnings = df.copy()
day_7_employee_earnings["earnings"] = day_7_earnings
day_7_employee_earnings
```

## Now writing our new dataframes in a parquet file format

```python
# Specify the Parquet file path
parquet_file_day_6 = './output_data/employee_earnings/earnings_date=2022-02-15/employee_earnings.parquet'

# Write the DataFrame to a Parquet file
day_6_employee_earnings.to_parquet(parquet_file_day_6)
```

```python
# Specify the Parquet file path
parquet_file_day_7 = './output_data/employee_earnings/earnings_date=2022-02-16/employee_earnings.parquet'

# Write the DataFrame to a Parquet file
day_7_employee_earnings.to_parquet(parquet_file_day_7)
```
